```
# basket = data.groupby(['InvoiceNo','Description'])['Quantity'].sum().unstack().fillna(0)
# basket


# basket = data.groupby(['InvoiceNo','Description'])['Quantity'].sum().unstack().fillna(0)


basket= data.groupby(['Transaction','Item'])['id'].sum().unstack().fillna(0)
basket
```

```
basket.tail(10)
```

```

basket=basket.applymap(lambda x:1 if x>0 else 0)
```

```

basket.tail(10)
```

```
popular_sets = apriori(basket, min_support=0.01, use_colnames=True)
```

```
popular_sets
```

```
popular_sets.tail(10)
```

```
rules = association_rules(popular_sets,metric='lift',min_threshold=1)
```

```

rules.tail(10)
```

```
rules.tail(50)
```

```
rules[ (rules['lift'] >= 1) &
       (rules['confidence'] >= 0.5) ]
```
